inspired by https://alanhylands.com/how-to-web-scrape-wikipedia-python-urllib-beautiful-soup-pandas/

```
import urllib.request
```

```
url = "https://en.wikipedia.org/wiki/2019%E2%80%9320_Premier_League"
```

```
page = urllib.request.urlopen(url)
```

```
from bs4 import BeautifulSoup
```

```
soup = BeautifulSoup(page, "lxml")
```

```
print(soup.prettify())
```

```
soup.title.string
```

```
all_tables=soup.find_all("table")
all_tables
```

```
# original -> right_table=soup.find('table', class_='wikitable sortable')
right_table = soup.find_all("table")[2]
right_table
```

```
A=[]
B=[]
C=[]
D=[]
E=[]
for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells):
        A.append(cells[0].find(text=True))
        mlnk=cells[1].findAll('a')
        B.append(mlnk[1].contents[0])
        clnk=cells[2].findAll('a')
        C.append(clnk[1].contents[0])
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
```

```
import pandas as pd
df=pd.DataFrame(A,columns=['Team'])
df['Manager']=B
df['Captain']=C
df['Kit Manufacturer']=D
df['Shirt Sponsor']=E
df
```

```
# clean up text that has \n at the end.

import re

p = re.compile(r'\n')
df['Team'] = [p.sub('', x) for x in df['Team']]
df['Kit Manufacturer'] = [p.sub('', x) for x in df['Kit Manufacturer']]
```
